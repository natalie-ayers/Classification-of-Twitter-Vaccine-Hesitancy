```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("tweets_by_day.csv")
df.loc[:, 'date'] = pd.to_datetime(df.loc[:, 'date'], format="%Y-%m-%d")
df.rename(columns={'num tweets': 'COVID Tweets', 'num vacc tweets': 'Vaccine Tweets', "date": "Date"}, inplace=True)
```

```python
sns.set(rc={'figure.figsize':(13, 6)})
plt_df = df.set_index('Date')
plt = sns.lineplot(data=plt_df)
plt.set_ylabel("Number of Tweets")
plt.title.set_text('Tweets by Day')
```
